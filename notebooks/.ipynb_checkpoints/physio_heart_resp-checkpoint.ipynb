```python
%load_ext autoreload
%autoreload 2
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as pl
import numpy as np
import glob
import os
from collections import defaultdict as dd
opj = os.path.join
from scipy.stats import sem, mode
from statsmodels.stats.anova import AnovaRM
```

```python
groups_dict = {'placebo':['sub-001_ses-1','sub-002_ses-2','sub-003_ses-3','sub-004_ses-2','sub-005_ses-1','sub-006_ses-3','sub-007_ses-3','sub-008_ses-1',
                         'sub-009_ses-2','sub-010_ses-1','sub-011_ses-3','sub-012_ses-2','sub-013_ses-3','sub-014_ses-2','sub-015_ses-2','sub-018_ses-1','sub-019_ses-2','sub-020_ses-1'],
               
              '5mg':['sub-001_ses-3','sub-002_ses-1','sub-003_ses-1','sub-004_ses-3','sub-005_ses-2','sub-006_ses-1','sub-007_ses-1','sub-008_ses-3',
                    'sub-009_ses-1','sub-010_ses-2','sub-011_ses-2','sub-012_ses-3','sub-013_ses-2','sub-014_ses-1','sub-015_ses-3','sub-018_ses-3','sub-019_ses-1','sub-020_ses-2'],
               
              '10mg':['sub-001_ses-2','sub-002_ses-3','sub-003_ses-2','sub-004_ses-1','sub-005_ses-3','sub-006_ses-2','sub-007_ses-2','sub-008_ses-2',
                     'sub-009_ses-3','sub-010_ses-3','sub-011_ses-1','sub-012_ses-1','sub-013_ses-1','sub-014_ses-3','sub-015_ses-1','sub-018_ses-2','sub-019_ses-3','sub-020_ses-3']}
```

```python
res_dict = dd(lambda:dd(lambda:dd(list)))


paths = glob.glob('/data1/projects/dumoulinlab/Lab_members/Marco/SM-pRF/sub-*/ses-*/physio/*task-2R_*_physio.csv')
paths = sorted([p for p in paths if 'ses-0' not in p and 'sim' not in p]) 

#pl.figure()
for path in paths:
    id = path.split(os.sep)[-1].split('_task')[0]

    df = pd.read_csv(path)

    hr_run_mean = df['hr'].mean()
    resp_run_mean = df['rvt'].mean()

    hr_run_var = df['hr'].var()
    resp_run_var = df['rvt'].var()

    for key in groups_dict.keys():
        if id in groups_dict[key]:
            res_dict[key][id]['heart rate'].append(hr_run_mean)
            res_dict[key][id]['resp rate'].append(resp_run_mean)
            res_dict[key][id]['heart rate var'].append(hr_run_var)
            res_dict[key][id]['resp rate var'].append(resp_run_var)
```

```python
pl.rcParams['axes.spines.right'] = False
pl.rcParams['axes.spines.top'] = False
pl.rcParams['ytick.labelsize'] = 22
pl.rcParams['xtick.labelsize'] = 22
pl.rcParams['axes.labelsize'] = 22


```

```python
pl.close('all')
fig,ax = pl.subplots(2,2,figsize=(16,16))
res_dict_sj = dd(lambda:dd(lambda:dd(list)))

for enum, key in enumerate(groups_dict.keys()):
    ids = sorted(res_dict[key].keys())
    print(ids)

    all_run_hr=[]
    all_run_rt=[]
    all_run_hr_sems=[]
    all_run_rt_sems=[]

    all_run_hr_var=[]
    all_run_rt_var=[]
    all_run_hr_var_sems=[]
    all_run_rt_var_sems=[]

    for inum,id in enumerate(ids):

        all_run_hr.append(np.mean(res_dict[key][id]['heart rate']))
        all_run_rt.append(np.mean(res_dict[key][id]['resp rate']))
        all_run_hr_sems.append(sem(res_dict[key][id]['heart rate']))
        all_run_rt_sems.append(sem(res_dict[key][id]['resp rate']))

        all_run_hr_var.append(np.mean(res_dict[key][id]['heart rate var']))
        all_run_rt_var.append(np.mean(res_dict[key][id]['resp rate var']))
        all_run_hr_var_sems.append(sem(res_dict[key][id]['heart rate var']))
        all_run_rt_var_sems.append(sem(res_dict[key][id]['resp rate var']))

        res_dict_sj[id.split('_')[0]][f'{key} heart rate mean'] = np.mean(res_dict[key][id]['heart rate'])
        res_dict_sj[id.split('_')[0]][f'{key} resp rate mean'] = np.mean(res_dict[key][id]['resp rate'])

        res_dict_sj[id.split('_')[0]][f'{key} heart rate var mean'] = np.mean(res_dict[key][id]['heart rate var'])
        res_dict_sj[id.split('_')[0]][f'{key} resp rate var mean'] = np.mean(res_dict[key][id]['resp rate var'])

        res_dict_sj[id.split('_')[0]][f'{key} xpos'] = np.linspace(enum-0.1,enum+0.1,len(ids))[inum]

        #if key == 'placebo':
        ax[0,0].text(res_dict_sj[id.split('_')[0]][f'{key} xpos'],res_dict_sj[id.split('_')[0]][f'{key} heart rate mean'],int(id.split('_')[0][3:]))
        ax[0,1].text(res_dict_sj[id.split('_')[0]][f'{key} xpos'],res_dict_sj[id.split('_')[0]][f'{key} resp rate mean'],int(id.split('_')[0][3:]))
        ax[1,0].text(res_dict_sj[id.split('_')[0]][f'{key} xpos'],res_dict_sj[id.split('_')[0]][f'{key} heart rate var mean'],int(id.split('_')[0][3:]))
        ax[1,1].text(res_dict_sj[id.split('_')[0]][f'{key} xpos'],res_dict_sj[id.split('_')[0]][f'{key} resp rate var mean'],int(id.split('_')[0][3:]))      

    ax[0,0].bar(enum,np.mean(all_run_hr))
    ax[0,1].bar(enum,np.mean(all_run_rt))
    ax[0,0].errorbar(np.linspace(enum-0.1,enum+0.1,len(all_run_hr)),all_run_hr,yerr=all_run_hr_sems,ls='',fmt='s',  mec='k', ecolor='k')    
    ax[0,1].errorbar(np.linspace(enum-0.1,enum+0.1,len(all_run_rt)),all_run_rt,yerr=all_run_rt_sems,ls='',fmt='s',  mec='k', ecolor='k')  

    ax[1,0].bar(enum,np.mean(all_run_hr_var))
    ax[1,1].bar(enum,np.mean(all_run_rt_var))
    ax[1,0].errorbar(np.linspace(enum-0.1,enum+0.1,len(all_run_hr_var)),all_run_hr_var,yerr=all_run_hr_var_sems,ls='',fmt='s',  mec='k', ecolor='k')    
    ax[1,1].errorbar(np.linspace(enum-0.1,enum+0.1,len(all_run_rt_var)),all_run_rt_var,yerr=all_run_rt_var_sems,ls='',fmt='s',  mec='k', ecolor='k')  

for sj in res_dict_sj.keys():
    try:
        ax[0,0].plot([res_dict_sj[sj][f'{key} xpos'] for key in ['placebo', '5mg', '10mg']], [res_dict_sj[sj][f'{key} heart rate mean'] for key in ['placebo', '5mg', '10mg']], ls='-', c='k')
        ax[0,1].plot([res_dict_sj[sj][f'{key} xpos'] for key in ['placebo', '5mg', '10mg']], [res_dict_sj[sj][f'{key} resp rate mean'] for key in ['placebo', '5mg', '10mg']], ls='-', c='k')
        ax[1,0].plot([res_dict_sj[sj][f'{key} xpos'] for key in ['placebo', '5mg', '10mg']], [res_dict_sj[sj][f'{key} heart rate var mean'] for key in ['placebo', '5mg', '10mg']], ls='-', c='k')
        ax[1,1].plot([res_dict_sj[sj][f'{key} xpos'] for key in ['placebo', '5mg', '10mg']], [res_dict_sj[sj][f'{key} resp rate var mean'] for key in ['placebo', '5mg', '10mg']], ls='-', c='k')
    except:
        pass

ax[0,0].set_xticks([0,1,2],['placebo','5mg','10mg'])
ax[0,1].set_xticks([0,1,2],['placebo','5mg','10mg'])
ax[0,0].set_ylabel('heart rate')
ax[0,1].set_ylabel('resp rate')

ax[1,0].set_xticks([0,1,2],['placebo','5mg','10mg'])
ax[1,1].set_xticks([0,1,2],['placebo','5mg','10mg'])
ax[1,0].set_ylabel('heart rate var')
ax[1,1].set_ylabel('resp rate var')
#pl.ylim(0.05,0.4) 11-3
```

```python
data=[]
for group in res_dict:
    for sj_id in res_dict[group]:
        for run in range(len(res_dict[group][sj_id]['heart rate'])):
        
            data.append([sj_id.split('_')[0], sj_id.split('_')[1], group, run, res_dict[group][sj_id]['heart rate'][run], res_dict[group][sj_id]['resp rate'][run], res_dict[group][sj_id]['heart rate var'][run], res_dict[group][sj_id]['resp rate var'][run]])


aa=pd.DataFrame(data, columns=['Subj', 'Ses', 'Dose', 'Run', 'heart rate', 'resp rate',  'heart rate var', 'resp rate var'])

anova_pvals = []

for sj in sorted(aa['Subj'].unique()):
    
    sj_dataset = aa[aa['Subj'] == sj]
    if sj == 'sub-004':
        sj_dataset = sj_dataset[sj_dataset['Run']<4]
    elif sj == 'sub-003':
        sj_dataset = sj_dataset[sj_dataset['Run']<2]
    elif sj == 'sub-007':
        sj_dataset = sj_dataset[sj_dataset['Run']<4]
    elif sj == 'sub-008':
        sj_dataset = sj_dataset[sj_dataset['Run']<5]
    elif sj == 'sub-009':
        sj_dataset = sj_dataset[sj_dataset['Run']<3]
    elif sj == 'sub-012':
        sj_dataset = sj_dataset[sj_dataset['Run']<4]
        
    print(sj)
    sj_res = [sj]

    aovrm2way = AnovaRM(sj_dataset, 'heart rate', 'Run', within=['Dose'])
    res2way = aovrm2way.fit()
    print('Dose effect on heart rate:')
    print(res2way)
    sj_res.append(res2way.anova_table['Pr > F']['Dose'])

    aovrm2way = AnovaRM(sj_dataset, 'resp rate', 'Run', within=['Dose'])
    res2way = aovrm2way.fit()
    print('Dose effect on resp rate:')
    print(res2way)
    sj_res.append(res2way.anova_table['Pr > F']['Dose'])

    aovrm2way = AnovaRM(sj_dataset, 'heart rate var', 'Run', within=['Dose'])
    res2way = aovrm2way.fit()
    print('Dose effect on heart rate var:')
    print(res2way)
    sj_res.append(res2way.anova_table['Pr > F']['Dose'])

    aovrm2way = AnovaRM(sj_dataset, 'resp rate var', 'Run', within=['Dose'])
    res2way = aovrm2way.fit()
    print('Dose effect on resp rate var:')
    print(res2way)
    sj_res.append(res2way.anova_table['Pr > F']['Dose'])

    print('\n')

    anova_pvals.append(sj_res)

```

```python
pd.options.display.float_format = '{:.5f}'.format

pd.DataFrame(anova_pvals, columns = ['Subj', 'heart rate Pval', 'resp rate Pval',  'heart rate var Pval', 'resp rate var Pval'])
```
