```python
res_dict = dd(lambda:dd(lambda:dd(list)))

paths = glob.glob('/data1/projects/dumoulinlab/Lab_members/Marco/SM-pRF/sourcedata/sub-*/*/task/*2R*/*task-2R*DotSwitchColorTimes.npy')
paths = sorted([p for p in paths if 'ses-0' not in p])# and 'sub-014' not in p and 'sub-007' not in p and 'sub-009' not in p]
#pl.figure()
for path in paths:
    id = path.split(os.sep)[-1].split('_task')[0]
    #print(id)
    aa = np.load(path)
    events_file = path.replace('DotSwitchColorTimes.npy','events.tsv')
    df = pd.read_csv(events_file, sep="\t")
    df_responses = df[df['event_type'] == 'response']
    #vals = []
    for bp in df_responses['onset']:
        diffs = bp-aa
        diffs_pos = diffs[diffs>0]
        if len(diffs_pos>0):
            val = np.min(diffs_pos)
            for key in groups_dict.keys():
                if id in groups_dict[key]:
                    res_dict[key][id]['resp times'].append(val)


```

```python
pl.figure(figsize=(14,5))
for key in groups_dict.keys():
    
    data = np.concatenate([res_dict[key][el]['resp times'] for el in res_dict[key]])
    y,x,_ = pl.hist(data,bins=200, label=key, alpha=0.8)
    pl.plot(np.median(data)*np.ones(10),np.linspace(0,y.max(),10), c='k',ls='--', label='median')
    pl.text(np.median(data), y.max(), f"{np.median(data):.2f}")
    pl.plot(np.mean(data)*np.ones(10),np.linspace(0,y.max(),10), c='r',ls='--', label='mean')
    pl.text(np.mean(data), y.max(), f"{np.mean(data):.2f}")
    pl.plot((x[np.argmax(y)]+0.5*(x[1]-x[0]))*np.ones(10),np.linspace(0,y.max(),10), c='w',ls='--', label='mode')
    pl.text((x[np.argmax(y)]+0.5*(x[1]-x[0])), y.max(), f"{(x[np.argmax(y)]+0.5*(x[1]-x[0])):.2f}")

    handles, labels = pl.gca().get_legend_handles_labels()

    # zip labels as keys and handles as values into a dictionary, ...
    # so only unique labels would be stored 
    dict_of_labels = dict(zip(labels, handles))

    # use unique labels (dict_of_labels.keys()) to generate your legend
    pl.legend(dict_of_labels.values(), dict_of_labels.keys())
    pl.xlabel('Time from dot color change (s)')
    pl.ylabel('Responses')
    pl.xlim(0,1.2)
```
