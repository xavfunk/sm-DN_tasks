```python
from scipy.stats import norm, mode
Z = norm.ppf

res_dict = dd(lambda:dd(lambda:dd(list)))



paths = glob.glob('/data1/projects/dumoulinlab/Lab_members/Marco/SM-pRF/sourcedata/sub-*/*/task/*2R*/*task-2R*DotSwitchColorTimes.npy')
paths = sorted([p for p in paths if 'ses-0' not in p and '2Rsim' not in p])# and 'sub-014' not in p and 'sub-007' not in p and 'sub-009' not in p]
#pl.figure()
for path in paths:
    id = path.split(os.sep)[-1].split('_task')[0]
    #print(id)
    aa = np.load(path)
    events_file = path.replace('DotSwitchColorTimes.npy','events.tsv')
    df = pd.read_csv(events_file, sep="\t")
    df_responses = df[df['event_type'] == 'response']
    vals = []
    for bp in df_responses['onset']:
        diffs = bp-aa
        diffs_pos = diffs[diffs>0]
        if len(diffs_pos>0):
            vals.append(np.min(diffs_pos))

    for key in groups_dict.keys():
        if id in groups_dict[key]:
            res_dict[key][id]['resp times run mean'].append(np.mean(vals))
            res_dict[key][id]['responses'].append(len(df_responses))


            y_hist,x,_ = pl.hist(vals,bins=np.linspace(0,3,300), label=key, alpha=0.0, density=True)

            kde_responses = gaussian_kde(vals,bw_method=x[5]-x[0])
            bins_kde = x

            y=kde_responses(x)

            res_dict[key][id]['resp times run mode'].append(x[np.argmax(y)])
            res_dict[key][id]['resp times y hist'].append(y_hist)
            res_dict[key][id]['resp times y kde'].append(y)
            res_dict[key][id]['resp times x hist'].append(x)

                    

    hits = 0
    false_alarms = 0
    for enum,dot_col_change_time in enumerate(aa):
        diffs = df_responses['onset'] - dot_col_change_time

        diffs_pos = diffs[diffs>0]
        if (enum+1)<len(aa):
            diffs_trial = diffs_pos[diffs_pos<(aa[enum+1]-dot_col_change_time)]
        else:
            diffs_trial = diffs_pos
           

        if len(diffs_trial>0):
            false_alarms += len(diffs_trial)-1
            if diffs_trial.min()<1.5:
                hits += 1
            else:
                false_alarms += 1


    for key in groups_dict.keys():
        if id in groups_dict[key]:
            if false_alarms == 0:
                false_alarms = 1
            res_dict[key][id]['false alarms run'].append(false_alarms)
            res_dict[key][id]['hits run'].append(hits)
            res_dict[key][id]['dprime'].append(Z(hits/len(aa))-Z(false_alarms/len(aa)))              


```

```python
data=[]
for group in res_dict:
    for sj_id in res_dict[group]:
        for run in range(len(res_dict[group][sj_id]['resp times run mean'])):
        
            data.append([sj_id.split('_')[0], sj_id.split('_')[1], group, run, res_dict[group][sj_id]['resp times run mean'][run], res_dict[group][sj_id]['dprime'][run]])

fig,ax = pl.subplots(1,2,figsize=(12,6))

aa=pd.DataFrame(data, columns=['Subj', 'Ses', 'Dose', 'Run', 'resp times run mean', 'dprime'])

anova_pvals = []

for enum,sj in enumerate(sorted(aa['Subj'].unique())):

    sj_res = [sj]
    
    sj_dataset = aa[aa['Subj'] == sj]
    print('\n')

    for en_pos,measure in enumerate(['resp times run mean','dprime']):

        aovrm2way = AnovaRM(sj_dataset, f"{measure}", 'Run', within=['Dose'])
        res2way = aovrm2way.fit()
        print(f'{sj} Dose effect on {measure}:')
        print(res2way)

        sj_res.append(res2way.anova_table['Pr > F']['Dose'])

        ax[en_pos].errorbar(enum, np.mean(sj_dataset[f"{measure}"]), sem(sj_dataset[f"{measure}"]), ls='',fmt='o',  mec='k', ecolor='k')
        ax[en_pos].text(enum, np.mean(sj_dataset[f"{measure}"]), int(sj[3:]))
        ax[en_pos].set_ylabel(f"{measure}")
        ax[en_pos].set_xticklabels([])

    anova_pvals.append(sj_res)
```
