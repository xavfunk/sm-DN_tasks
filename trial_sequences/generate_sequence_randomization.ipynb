```python
import random

# Define parameters
num_subjects = 20
num_sessions = 3
num_items = 18
items_per_session = 6
target_total_per_item = num_subjects  # Each item should appear 20 times total
target_per_session = target_total_per_item // num_sessions  # Roughly 7 per session

# Initialize counters
total_item_counts = {item: 0 for item in range(num_items)}
session_item_counts = [{item: 0 for item in range(num_items)} for _ in range(num_sessions)]

# Initialize storage for results
subjects_sessions = []

# Assign items per subject and session
for subject in range(num_subjects):
    subject_sessions = []
    used_items = set()  # Track items already used by the subject

    for session in range(num_sessions):
        session_items = []
        while len(session_items) < items_per_session:
            item = random.choice(range(num_items))

            # Check if the item can be assigned without breaking balance
            if (total_item_counts[item] < target_total_per_item and
                session_item_counts[session][item] < target_per_session + 1 and  # Allow some leeway
                item not in used_items):

                # Assign item
                session_items.append(item)
                used_items.add(item)
                total_item_counts[item] += 1
                session_item_counts[session][item] += 1

        subject_sessions.append(session_items)

    subjects_sessions.append(subject_sessions)

# Display the result for each subject
for subject, sessions in enumerate(subjects_sessions):
    print(f"Subject {subject + 1}:")
    for session, items in enumerate(sessions):
        print(f"  Session {session + 1}: Items {items}")
```
